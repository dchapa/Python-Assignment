## Research Question
What are the main components of the long term healthcare costs in the Netherlands and how have they changed in past 15 years before 2013?

#### Motivation
The healthcare costs are a substantial part of the government budget and have a big impact on the costs of individuals Especially with the increasing costs of it over the years, it is interesting to study the main determinants of the long term healthcare costs in the Netherlands. 

According to the [OECD Health Statistics](http://www.oecd.org/els/health-systems/Briefing-Note-NETHERLANDS-2014.pdf), the healthcare costs in the Netherlands are about 11.8% of the GDP in 2012, and of all the OECD countries only the United States ranks higher with a spending of 16.9%. 86% of the healthcare expenses in the Netherlands are paid by the government or mandatory social insurance, which is notably different compared to the average of 72% of the OECD countries in which the individuals bear more risks with private insurance. 

#### Method
To answer our research question we will try to explain what the main determinants are of the increasing healthcare costs with the use of data obtained from the [Centraal Planbureau](https://www.cpb.nl/). Based on this data we will explain how the funding of healthcare costs changed over the years in the period 1998 till 2013.

#### Data
The data gives information about the healthcare expenses in the Netherlands over the period from 1998 till 2013, and how these costs are divided over several financing resources. The main determinants of healthcare can be grouped in impact of demography and epidemiology, impact of structural growth of GDP, relative prices in care and other growth in the economy. 

During the past decade the population has been getting older and healthier over time. Higher life expectancy at birth can increase the costs of care, while if the life expectancy in good health increases people will work more years, and this increases the funding of healthcare.  

As well the impact of the structural growth of GDP per capita influences the costs of care. Several studies by the [WHO](http://www.who.int/health_financing/documents/report_en_11_deter-he.pdf) have shown that GDP per capita can be identified as a very important factor for explaining differences across countries in the level and growth of total healthcare expenditures. According to the [CPB Report]((https://www.cpb.nl/publicatie/middellangetermijnverkenning-2018-2021) the income elasticity of healthcare expenditures is 0.6, which means that if the GDP per capita grows with 1%, then the healthcare expenses will grow with 0.6%.
